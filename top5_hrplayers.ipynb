```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
judge_2022 = pd.read_csv('judge_2022_daily_stats.csv')
riley_2022 = pd.read_csv('riley_2022_daily_stats.csv')
alonso_2022 = pd.read_csv('alonso_2022_daily_stats.csv')
schwarber_2022 = pd.read_csv('schwarber_2022_daily_stats.csv')
trout_2022 = pd.read_csv('trout_2022_daily_stats.csv')
judge_2022.head()
```

```python
judge_2022['Month'] = pd.to_datetime(judge_2022['Date']).dt.month
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August',
               9: 'September', 10: 'October'}
judge_2022['Month'] = judge_2022['Month'].map(month_names)
month_order = ['April', 'May', 'June', 'July', 'August', 'September', 'October']
judge_2022['Month'] = pd.Categorical(judge_2022['Month'], categories=month_order, ordered=True)

# Create a pivot table to calculate the number of home runs for each day of the week and month
pivot_table = judge_2022.pivot_table(index='day_of_week', columns='Month', values='HR', aggfunc='sum')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='PuRd',  cbar=False)

# Set the title and labels
plt.title('Aaron Judge Home Runs by Day and Month')
plt.xlabel('Month')
plt.ylabel('Day of Week')

# Display the heatmap
plt.show()
```

```python
judge_2022['Month'] = pd.to_datetime(judge_2022['Date']).dt.month
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August',
               9: 'September', 10: 'October'}
judge_2022['Month'] = judge_2022['Month'].map(month_names)
month_order = ['April', 'May', 'June', 'July', 'August', 'September', 'October']
judge_2022['Month'] = pd.Categorical(judge_2022['Month'], categories=month_order, ordered=True)

# Create a pivot table to calculate the number of home runs for each day of the week and month
pivot_table = judge_2022.pivot_table(index='day_of_week', columns='Month', values='AVG', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='.3f', cmap='PuRd', cbar=True)

# Set the title and labels
plt.title('Aaron Judge AVG by Day and Month')
plt.xlabel('Month')
plt.ylabel('Day of Week')

# Display the heatmap
plt.show()
```
